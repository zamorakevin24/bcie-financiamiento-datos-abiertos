```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
df_aprobaciones = pd.read_csv('aprobaciones_limpias.csv')
df_desembolsos = pd.read_csv('desembolsos_limpios.csv')
```

```python
df_aprobaciones.head()
```

```python
monto_anio_aprob = df_aprobaciones.groupby('ANIO_APROBACION').sum('MONTO_BRUTO_USD')['MONTO_BRUTO_USD']
```

```python
df_monto_aprob = pd.DataFrame(monto_anio_aprob,columns=['MONTO_BRUTO_USD'])
```

```python
df_monto_aprob = df_monto_aprob.reset_index()
```

```python
df_monto_aprob
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(df_monto_aprob["ANIO_APROBACION"], df_monto_aprob["MONTO_BRUTO_USD"], marker="o") 
plt.xlabel("Año de aprobación") 
plt.ylabel("Monto aprobado (miles de millones de USD)") 
plt.title("Evolución del monto aprobado por el BCIE (1961–2025)") 
plt.grid(True) 
plt.savefig("Graficos/01_monto_aprobado_por_anio.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_desembolsos.head()
```

```python
monto_anio_des = df_desembolsos.groupby('ANIO_DESEMBOLSO').sum('MONTO_BRUTO_USD')['MONTO_BRUTO_USD']
df_monto_des = pd.DataFrame(monto_anio_des,columns=['MONTO_BRUTO_USD'])
df_monto_des = df_monto_des.reset_index()
df_monto_des
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(df_monto_des["ANIO_DESEMBOLSO"], df_monto_des["MONTO_BRUTO_USD"], marker="o") 
plt.xlabel("Año de desembolso") 
plt.ylabel("Monto desembolsado (miles de millones de USD)") 
plt.title("Evolución del monto desembolsado por el BCIE (1964–2025)") 
plt.grid(True) 
plt.savefig("Graficos/02_monto_desembolso_por_año.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df_desembolsos.head()
```

```python
df_desem_decada = df_desembolsos.groupby('DECADA').sum('MONTO_BRUTO_USD')['MONTO_BRUTO_USD']
df_desem_decada = pd.DataFrame(df_desem_decada)
df_desem_decada['MONTO_APROBADO'] = df_desem_decada['MONTO_BRUTO_USD']
df_desem_decada = df_desem_decada.reset_index()
df_desem_decada
```

```python
df_desem_decada = df_desem_decada.drop(columns=['MONTO_BRUTO_USD'])
```

```python
df_desem_decada
```

```python
df_aprobaciones
```

```python
df_aprob_decada = df_aprobaciones.groupby('DECADA').sum('MONTO_BRUTO_USD')['MONTO_BRUTO_USD']
df_aprob_decada = pd.DataFrame(df_aprob_decada)
```

```python
df_aprob_decada = df_aprob_decada.reset_index()
```

```python
df_aprob_decada['MONTO_DESEMBOLSADO'] = df_aprob_decada['MONTO_BRUTO_USD']
```

```python
df_aprob_decada = df_aprob_decada.drop(columns=['MONTO_BRUTO_USD'])
```

```python
df_aprob_decada
```

```python
df_desem_decada
```

```python
df_comparacion = pd.merge(df_aprob_decada,df_desem_decada,on='DECADA')
```

```python
df_comparacion
```

```python
df_comparacion["APROBADO_B"] = df_comparacion["MONTO_APROBADO"] / 1e9
df_comparacion["DESEMBOLSADO_B"] = df_comparacion["MONTO_DESEMBOLSADO"] / 1e9
```

```python
df_comparacion_plot = df_comparacion.drop(columns=['MONTO_DESEMBOLSADO','MONTO_APROBADO'])
```

```python
df_plot = df_comparacion_plot.sort_values('DECADA')
```

```python
df_plot
```

```python
x = np.arange(len(df_plot))
```

```python
x
```

```python
width = 0.35
```

```python
plt.figure(figsize=(10,6))

# Barras de Aprobaciones (a la izquierda)
plt.bar(
    x - width/2,
    df_plot["APROBADO_B"],
    width,
    label="Aprobado"
)

# Barras de Desembolsos (a la derecha)
plt.bar(
    x + width/2,
    df_plot["DESEMBOLSADO_B"],
    width,
    label="Desembolsado"
)

# Etiquetas del eje X (décadas)
plt.xticks(x, df_plot["DECADA"])

plt.xlabel("Década")
plt.ylabel("Monto (miles de millones de USD)")
plt.title("Aprobaciones vs Desembolsos por década (BCIE, 1960–2025)")

plt.legend()
plt.grid(axis="y", alpha=0.3)

plt.tight_layout()

plt.savefig(
    "Graficos/03_aprob_vs_desemb_por_decada.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
